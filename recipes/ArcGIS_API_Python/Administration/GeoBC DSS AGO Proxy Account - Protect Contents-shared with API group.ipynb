### Import modules

```python
import arcgis
from getpass import getpass
```

### Connect AGO Account

```python
url = 'http://www.arcgis.com'
username = 'PX.GeoBC.DSS.Creator'


password =getpass()

gis = arcgis.GIS(url, username, password)
```

```python
# Check to make sure it connects the account
user_test = gis.users.me
user_test
```

## <font color='red'>Search all content items owned by the account.

```python
DSS_content = gis.content.search(query="owner:" + gis.users.me.username, 
                                #item_type="Feature Layer", 
                                max_items=5000)
```

## loop the search result and turn on item 'Delete protection' button for all items

```python
for DSS_item in DSS_content:
    if DSS_item.protected == True:
        print (DSS_item,' already protected.')
        
    else:
        DSS_item.protect(enable = True)
        print (DSS_item, " Item delete protection is turned on.")
        

print ('All items are protected from being accidentally deleted.')
```

## Search and list all groups owned  by the account

```python
DSS_group = gis.groups.search(query="owner:" + gis.users.me.username)
DSS_group
```

## loop through the search result and turn on Group 'Delete protection' button

```python
for gp in DSS_group:
    if gp.protected == True:
        print (gp,' already protected')
        
    else:
        gp.protected = True
        print (gp, " Group delete protection is turned on.")
        

print ('All groups are protected from being accidentally deleted.')
```
